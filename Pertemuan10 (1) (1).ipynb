### Analisis

Program ini menggunakan perulangan while dengan i mulai dari 1 hingga 13, bertambah 2 setiap iterasi. Jika i sama dengan 7, program mencetak "WOW", selain itu mencetak nilai i.

```python
i = 1
while i <= 8:
    if i % 2 == 1:
        print("hidup")
    else:
        print("mahasiswa")
    i += 1
```
